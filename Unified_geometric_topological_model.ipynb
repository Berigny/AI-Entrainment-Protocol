```python
import numpy as np
import matplotlib.pyplot as plt
import mne  # Main EEG processing library
from scipy import signal, stats
import antropy as ant  # For entropy calculation
```

```python
# Load a sample dataset from MNE
from mne.datasets import eegbci
from mne import io

# This downloads the BCI dataset. Replace with your data path.
data_path = eegbci.load_data(1, 1)  # Subject 1, run 1
raw = io.read_raw_edf(data_path[0], preload=True)
```

```python
# 1. Filter the data to remove slow drifts and line noise
raw.filter(1, 40)  # Bandpass filter between 1-40 Hz

# 2. Notch filter to remove 60 Hz (or 50 Hz) power line interference
raw.notch_filter(60) # Use 50 for European countries

# 3. Select channels of interest (e.g., frontal and central)
# The model predicts global coherence, so many channels are relevant.
picks = mne.pick_types(raw.info, eeg=True) # Pick all EEG channels

# 4. For entropy analysis, we often use a single representative channel.
# Let's use a central channel like 'Cz' if available.
if 'Cz' in raw.ch_names:
    chan_of_interest = 'Cz'
else:
    chan_of_interest = raw.ch_names[0] # Or pick another
```

```python
# Define events or time periods for different states.
# Example: First 60s are rest, next 60s are task.
# You will need to define this based on your experiment's timeline.

# For a simple test, create artificial events
sfreq = raw.info['sfreq']
duration = int(60 * sfreq)  # 60 seconds in samples, convert to integer

# Create epochs for "Rest" state (first 60 seconds)
data_rest = raw.get_data(picks=chan_of_interest, start=0, stop=duration)[0]

# Create epochs for "Task" state (next 60 seconds)
data_task = raw.get_data(picks=chan_of_interest, start=duration, stop=2*duration)[0]

# For a more complex analysis with real event markers, use mne.Epochs.
```

```python
# Calculate Power Spectral Density (PSD) for the rest state
freqs, psd = signal.welch(data_rest, fs=sfreq, nperseg=2*sfreq)

# Plot the PSD
plt.figure(figsize=(10, 5))
plt.plot(freqs, psd, color='k', lw=2)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power Spectral Density (V²/Hz)')
plt.title('EEG Power Spectrum during Rest')
plt.xlim(1, 40)
plt.axvline(11, color='r', linestyle='--', alpha=0.5, label='11 Hz Prediction')
plt.axvline(13, color='g', linestyle='--', alpha=0.5, label='13 Hz Prediction')
plt.axvline(22, color='b', linestyle='--', alpha=0.5, label='22 Hz Prediction')
plt.legend()
plt.show()

# Check if peaks exist near predicted frequencies
def find_peak_near_freq(freq_array, power_array, target_freq, tolerance=1.0):
    idx = np.where((freq_array > target_freq - tolerance) & (freq_array < target_freq + tolerance))[0]
    if len(idx) > 0:
        local_max_idx = idx[np.argmax(power_array[idx])]
        return freqs[local_max_idx], psd[local_max_idx]
    else:
        return None, None

for target_freq in [11, 13, 22]:
    peak_freq, peak_power = find_peak_near_freq(freqs, psd, target_freq)
    if peak_freq:
        print(f"Found peak at {peak_freq:.2f} Hz near target {target_freq} Hz.")
    else:
        print(f"No significant peak found near {target_freq} Hz.")

# (Optional) Calculate coherence between channel pairs
# This is more complex but powerful evidence for the model.
```

```python
# Calculate Sample Entropy for both conditions
sampen_rest = ant.sample_entropy(data_rest, order=2, metric='chebyshev')
sampen_task = ant.sample_entropy(data_task, order=2, metric='chebyshev')

print(f"Sample Entropy - Rest: {sampen_rest:.4f}")
print(f"Sample Entropy - Task: {sampen_task:.4f}")

# The model predicts sampen_rest > sampen_task
if sampen_rest > sampen_task:
    print("Result supports Prediction 6: Entropy is higher during rest.")
else:
    print("Result does not support the prediction.")

# For robustness, calculate another entropy measure (e.g., Spectral Entropy)
spen_rest = ant.spectral_entropy(data_rest, sfreq, method='welch', normalize=True)
spen_task = ant.spectral_entropy(data_task, sfreq, method='welch', normalize=True)

print(f"Spectral Entropy - Rest: {spen_rest:.4f}")
print(f"Spectral Entropy - Task: {spen_task:.4f}")
```

```python
# If you have multiple epochs, you can run a statistical test.
# This example uses a simple t-test on entropy values from two conditions.

# Let's simulate multiple epochs by splitting the data
n_epochs = 10
rest_epochs = np.array_split(data_rest, n_epochs)
task_epochs = np.array_split(data_task, n_epochs)

# Calculate entropy for each epoch
sampen_rest_vals = [ant.sample_entropy(epoch) for epoch in rest_epochs]
sampen_task_vals = [ant.sample_entropy(epoch) for epoch in task_epochs]

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(sampen_rest_vals, sampen_task_vals)

print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.3f}")
if p_value < 0.05:
    print("The difference in entropy between conditions is statistically significant.")
else:
    print("No significant difference found.")
```

```python
# Define the length of an epoch (e.g., 4 seconds is standard)
epoch_length = 4 * int(raw.info['sfreq']) # 4 seconds * sampling rate

# Create arrays to store entropy values for MANY epochs
sampen_rest_vals = []
sampen_task_vals = []

# Segment the entire "rest" data into 4-second epochs and calculate entropy for each
for i in range(0, len(data_rest), epoch_length):
    epoch = data_rest[i:i+epoch_length]
    if len(epoch) == epoch_length: # Ensure it's a full epoch
        sampen_rest_vals.append(ant.sample_entropy(epoch))

# Do the same for the "task" data
for i in range(0, len(data_task), epoch_length):
    epoch = data_task[i:i+epoch_length]
    if len(epoch) == epoch_length:
        sampen_task_vals.append(ant.sample_entropy(epoch))

# Now perform a proper statistical test
from scipy import stats
t_stat, p_value = stats.ttest_ind(sampen_rest_vals, sampen_task_vals)

print(f"Mean Sample Entropy - Rest: {np.mean(sampen_rest_vals):.4f}")
print(f"Mean Sample Entropy - Task: {np.mean(sampen_task_vals):.4f}")
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.3e}") # Use exponential notation for small p-values

# Plot the results
plt.figure(figsize=(8, 6))
plt.boxplot([sampen_rest_vals, sampen_task_vals], labels=['Rest', 'Task'])
plt.title('Sample Entropy Across Multiple Epochs')
plt.ylabel('Sample Entropy')
plt.grid(True)
plt.show()
```

```python
# Install necessary libraries
!pip install mne antropy

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import mne
from scipy import signal, stats
import antropy as ant
from mne.datasets import eegbci

# --- Configuration ---
subject = 1
runs = [1, 2, 3, 4, 5]  # Analyze runs 1 through 5
epoch_duration = 4  # seconds

# Initialize lists to store results from ALL runs
all_sampen_rest = []
all_sampen_task = []
all_psd_rest_epochs = []  # We'll store PSDs for grand average later

# Loop through each run
for run in runs:
    print(f"\n=== Processing Subject {subject}, Run {run} ===\n")

    # 1. Load the data for this run
    data_path = eegbci.load_data(subject, run, update_path=True)
    raw = mne.io.read_raw_edf(data_path[0], preload=True)

    # 2. Preprocess the data
    raw.filter(1, 40.)
    raw.notch_filter(60.)
    sfreq = raw.info['sfreq']
    picks = mne.pick_types(raw.info, eeg=True)

    # 3. Define conditions (Assuming first 60s are rest, next 60s are task)
    duration_samples = 60 * int(sfreq)
    # Use the first available channel instead of 'Cz'
    channel_to_analyze = raw.ch_names[0]

    # Ensure stop times do not exceed the actual duration of the recording
    rest_stop_samples = min(duration_samples, len(raw.times))
    task_start_samples = duration_samples
    task_stop_samples = min(2 * duration_samples, len(raw.times))

    data_rest = raw.get_data(picks=channel_to_analyze, start=0, stop=rest_stop_samples)[0]
    data_task = raw.get_data(picks=channel_to_analyze, start=task_start_samples, stop=task_stop_samples)[0]


    # 4. --- Analysis for Prediction 6 (Entropy) ---
    epoch_samples = epoch_duration * int(sfreq)
    sampen_rest_vals = []
    sampen_task_vals = []

    # Segment data into epochs and calculate entropy
    for i in range(0, len(data_rest), epoch_samples):
        epoch = data_rest[i:i+epoch_samples]
        if len(epoch) == epoch_samples:
            sampen_rest_vals.append(ant.sample_entropy(epoch))
    for i in range(0, len(data_task), epoch_samples):
        epoch = data_task[i:i+epoch_samples]
        if len(epoch) == epoch_samples:
            sampen_task_vals.append(ant.sample_entropy(epoch))

    # Store results from this run
    all_sampen_rest.extend(sampen_rest_vals)
    all_sampen_task.extend(sampen_task_vals)

    # Print results for this individual run
    print(f"Run {run} - Mean Sample Entropy:")
    if sampen_rest_vals: # Check if list is not empty before calculating mean
        print(f"  Rest: {np.mean(sampen_rest_vals):.4f}")
    else:
        print("  Rest: Not enough data for epochs")
    if sampen_task_vals: # Check if list is not empty
         print(f"  Task: {np.mean(sampen_task_vals):.4f}")
    else:
        print("  Task: Not enough data for epochs")

    if sampen_rest_vals and sampen_task_vals: # Only perform t-test if both lists have data
        t_stat, p_value = stats.ttest_ind(sampen_rest_vals, sampen_task_vals)
        print(f"  T-test: p = {p_value:.4f}")
    else:
        print("  T-test: Not enough data in one or both conditions")


    # 5. --- (Optional) Store PSD data for Prediction 3 ---
    # Let's calculate the PSD for the entire rest block for a grand average later
    if len(data_rest) > 0: # Only calculate PSD if there is rest data
        freqs, psd_rest = signal.welch(data_rest, fs=sfreq, nperseg=min(len(data_rest), 2*int(sfreq)))
        all_psd_rest_epochs.append(psd_rest) # Store the PSD curve for this run
    else:
        print("  Not enough rest data to calculate PSD.")


# --- GRAND ANALYSIS AFTER PROCESSING ALL 5 RUNS ---

print(f"\n\n=== GRAND RESULTS (Across {len(runs)} Runs) ===")

# 6. Analyze Aggregate Entropy Data (Prediction 6)
print(f"\n--- Prediction 6: Entropy Analysis ---")
print(f"Total Epochs Analyzed: Rest: {len(all_sampen_rest)}, Task: {len(all_sampen_task)}")
print(f"Grand Mean Sample Entropy:")
if all_sampen_rest:
    print(f"  Rest: {np.mean(all_sampen_rest):.4f}")
else:
    print("  Rest: Not enough data for grand mean")
if all_sampen_task:
    print(f"  Task: {np.mean(all_sampen_task):.4f}")
else:
    print("  Task: Not enough data for grand mean")


# Perform a massive t-test on all epochs from all runs
if all_sampen_rest and all_sampen_task:
    t_stat_grand, p_value_grand = stats.ttest_ind(all_sampen_rest, all_sampen_task)
    print(f"Grand T-test: t = {t_stat_grand:.3f}, p = {p_value_grand:.3e}")
else:
    print("Grand T-test: Not enough data in one or both conditions for grand test")


# Plot the grand boxplot
if all_sampen_rest or all_sampen_task: # Only plot if there is data in at least one condition
    plt.figure(figsize=(8, 6))
    # Handle cases where one list might be empty for plotting
    plot_data = [all_sampen_rest] if all_sampen_rest else []
    plot_labels = ['Rest'] if all_sampen_rest else []
    if all_sampen_task:
        plot_data.append(all_sampen_task)
        plot_labels.append('Task')

    if plot_data: # Ensure there is data to plot
        plt.boxplot(plot_data, labels=plot_labels)
        plt.title(f'Sample Entropy - Grand Results (Across {len(runs)} Runs)')
        plt.ylabel('Sample Entropy')
        plt.grid(True)
        plt.show()
    else:
        print("Not enough data in either condition to generate boxplot.")
else:
    print("Not enough data in either condition to generate boxplot.")


# 7. Analyze Aggregate Spectral Data (Prediction 3)
print(f"\n--- Prediction 3: Spectral Analysis ---")
# Create a grand average PSD across all 5 runs
if all_psd_rest_epochs: # Only calculate grand average PSD if there are PSDs stored
    grand_avg_psd = np.mean(all_psd_rest_epochs, axis=0)
    grand_std_psd = np.std(all_psd_rest_epochs, axis=0)
    grand_se_psd = grand_std_psd / np.sqrt(len(all_psd_rest_epochs))

    # Plot the Grand Average PSD with confidence interval
    plt.figure(figsize=(12, 5))
    plt.plot(freqs, grand_avg_psd, color='k', lw=2, label='Grand Mean PSD')
    plt.fill_between(freqs, grand_avg_psd - grand_se_psd, grand_avg_psd + grand_se_psd, color='gray', alpha=0.5, label='±1 SE')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Power Spectral Density (V²/Hz)')
    plt.title(f'Grand Average EEG Power Spectrum during Rest (Across {len(runs)} Runs)')
    plt.xlim(1, 40)
    # plt.ylim(0, 20) # You may need to adjust this
    for target_freq in [11, 13, 22]:
        plt.axvline(target_freq, color='r', linestyle='--', alpha=0.5, label=f'Predicted: {target_freq} Hz' if target_freq == 11 else "")
    plt.legend()
    plt.show()
else:
    print("Not enough rest data to calculate and plot Grand Average PSD.")


# 8. (Optional) Grand Coherence Analysis could be added here in a similar way.
```

```python
# @title Diagram 1: Interactive Star Tetrahedron (Plotly) with Cube Edges + S1/S2 Labels + IC/EC
import numpy as np
import itertools
import plotly.graph_objects as go

# ---------- Geometry helpers ----------
def cube_vertices(a=1.0):
    """All 8 vertices of a cube centred at the origin with half-edge length a."""
    signs = [-a, a]
    return np.array([[x, y, z] for x in signs for y in signs for z in signs], dtype=float)

def cube_edges(verts, a=1.0):
    """Pairs of indices for cube edges (vertices differ in exactly one coordinate)."""
    edges = []
    n = len(verts)
    for i in range(n):
        for j in range(i+1, n):
            diff = np.abs(verts[i] - verts[j])
            # Edge: exactly one coordinate differs by 2a, others equal
            if np.count_nonzero(diff) == 1 and np.isclose(diff.max(), 2*a):
                edges.append((i, j))
    return edges

def star_tetra_partitions(verts):
    """
    Partition cube vertices into two interpenetrating tetrahedra.
    Use sign(x*y*z): +1 => Tetra A, -1 => Tetra B.
    """
    prod = np.sign(verts[:, 0] * verts[:, 1] * verts[:, 2])
    idx_A = np.where(prod > 0)[0]  # Tetra A
    idx_B = np.where(prod < 0)[0]  # Tetra B
    return idx_A, idx_B

def tetra_edges(indices):
    """All 6 edges of a tetrahedron (K4) from its 4 vertex indices."""
    return list(itertools.combinations(indices, 2))

def segments_to_lines(verts, edges):
    """Convert edge index pairs to x/y/z lists with None separators for Plotly line traces."""
    xs, ys, zs = [], [], []
    for i, j in edges:
        xs += [verts[i, 0], verts[j, 0], None]
        ys += [verts[i, 1], verts[j, 1], None]
        zs += [verts[i, 2], verts[j, 2], None]
    return xs, ys, zs

# ---------- Build shape ----------
a = 1.0
V = cube_vertices(a=a)
E_cube = cube_edges(V, a=a)
idxA, idxB = star_tetra_partitions(V)
E_tetA = tetra_edges(idxA)
E_tetB = tetra_edges(idxB)

# Assign labels (S1 on one tetra, S2 on the other)
labels_S1 = ["S1-N0 (Novelty)", "S1-N1 (Uniqueness)", "S1-N2 (Connection)", "S1-N3 (Action)"]
labels_S2 = ["S2-N0 (Imagination)", "S2-N1 (Autonomy)", "S2-N2 (Relatedness)", "S2-N3 (Mastery)"]

# Sort index arrays for a stable mapping
idxA_sorted = sorted(idxA.tolist())
idxB_sorted = sorted(idxB.tolist())

label_by_index = {}
for k, vi in enumerate(idxA_sorted):
    label_by_index[vi] = labels_S1[k]
for k, vi in enumerate(idxB_sorted):
    label_by_index[vi] = labels_S2[k]

# Fallback text for safety
vertex_text = [label_by_index.get(i, f"v{i}") for i in range(len(V))]

# Detached nodes for IC and EC (above/below the shape on Z)
IC = np.array([[0.0, 0.0, 1.8*a]])
EC = np.array([[0.0, 0.0, -1.8*a]])

# ---------- Traces ----------
# Cube edges (context)
cx, cy, cz = segments_to_lines(V, E_cube)
tr_cube = go.Scatter3d(
    x=cx, y=cy, z=cz,
    mode="lines",
    line=dict(width=2),
    opacity=0.35,
    name="Cube edges",
    hoverinfo="skip",
    showlegend=True
)

# Tetra A edges
ax_, ay_, az_ = segments_to_lines(V, E_tetA)
tr_tetA = go.Scatter3d(
    x=ax_, y=ay_, z=az_,
    mode="lines",
    line=dict(width=5),
    name="Tetrahedron A (S1)",
    showlegend=True
)

# Tetra B edges
bx_, by_, bz_ = segments_to_lines(V, E_tetB)
tr_tetB = go.Scatter3d(
    x=bx_, y=by_, z=bz_,
    mode="lines",
    line=dict(width=5),
    name="Tetrahedron B (S2)",
    showlegend=True
)

# All cube vertices with labels (hover + on-plot text)
tr_vertices = go.Scatter3d(
    x=V[:, 0], y=V[:, 1], z=V[:, 2],
    mode="markers+text",
    text=vertex_text,
    textposition="top center",
    marker=dict(size=4),
    name="S1/S2 vertices",
    hovertemplate="%{text}<br>(%{x}, %{y}, %{z})<extra></extra>",
    showlegend=True
)

# Detached IC / EC nodes
IC = np.array([[0.0, 0.0, 0.0]])

tr_ic = go.Scatter3d(
    x=[IC[0, 0]], y=[IC[0, 1]], z=[IC[0, 2]],
    mode="markers+text",
    text=["IC"],
    textposition="top center",
    marker=dict(size=8, color="orange", symbol="circle"),
    name="IC (Internal Coherence)",
    hovertemplate="IC (Internal Coherence)<extra></extra>",
    showlegend=True
)
tr_ec = go.Scatter3d(
    x=EC[:, 0], y=EC[:, 1], z=EC[:, 2],
    mode="markers+text",
    text=["EC"],
    textposition="top center",
    marker=dict(size=6),
    name="EC (External Coherence)",
    hovertemplate="EC (External Coherence)<extra></extra>",
    showlegend=True
)

# ---------- Layout ----------
layout = go.Layout(
    title="Star Tetrahedron (Stella Octangula) within Cube • S1/S2 Labeled + IC/EC",
    scene=dict(
        xaxis=dict(visible=False),
        yaxis=dict(visible=False),
        zaxis=dict(visible=False),
        aspectmode="data",
    ),
    legend=dict(x=0.02, y=0.98),
    margin=dict(l=0, r=0, t=60, b=0)
)

fig = go.Figure(data=[tr_cube, tr_tetA, tr_tetB, tr_vertices, tr_ic, tr_ec], layout=layout)
fig.show()

# Optional: print the mapping for reference in logs
print("Vertex → Label mapping:")
for i in range(len(V)):
    print(f"{i}: {vertex_text[i]}   coords={tuple(V[i])}")
print("IC at", tuple(IC[0]), "| EC at", tuple(EC[0]))
```

```python
# @title Diagram 3. Interactive 3D Packet Flow over Star Tetrahedron (with IC bridge & loop) ---
import numpy as np
import itertools
import plotly.graph_objects as go

# =========================
# Geometry helpers
# =========================
def cube_vertices(a=1.0):
    signs = [-a, a]
    return np.array([[x, y, z] for x in signs for y in signs for z in signs], dtype=float)

def cube_edges(verts, a=1.0):
    edges = []
    n = len(verts)
    for i in range(n):
        for j in range(i+1, n):
            diff = np.abs(verts[i] - verts[j])
            if np.count_nonzero(diff) == 1 and np.isclose(diff.max(), 2*a):
                edges.append((i, j))
    return edges

def star_tetra_partitions(verts):
    prod = np.sign(verts[:,0] * verts[:,1] * verts[:,2])
    idx_A = np.where(prod > 0)[0]  # S1
    idx_B = np.where(prod < 0)[0]  # S2
    return idx_A.tolist(), idx_B.tolist()

def tetra_edges(indices):
    return list(itertools.combinations(indices, 2))  # K4 edges (6)

def segments_to_lines(verts, edges):
    xs, ys, zs = [], [], []
    for i, j in edges:
        xs += [verts[i,0], verts[j,0], None]
        ys += [verts[i,1], verts[j,1], None]
        zs += [verts[i,2], verts[j,2], None]
    return xs, ys, zs

# Interpolate a polyline segment between two points with n steps
def interp_segment(p0, p1, n=25):
    t = np.linspace(0, 1, n, endpoint=False)  # exclude endpoint to avoid duplicate vertices
    return p0[None,:] * (1-t[:,None]) + p1[None,:] * t[:,None]

# =========================
# Build geometry
# =========================
a = 1.0
V = cube_vertices(a=a)
E_cube = cube_edges(V, a=a)
idxA, idxB = star_tetra_partitions(V)
E_A = tetra_edges(idxA)
E_B = tetra_edges(idxB)
IC = np.array([0.0, 0.0, 0.0])

# Labels (stable order for display)
labels_S1 = ["S1-N0 (Novelty)", "S1-N1 (Uniqueness)", "S1-N2 (Connection)", "S1-N3 (Action)"]
labels_S2 = ["S2-N0 (Imagination)", "S2-N1 (Autonomy)", "S2-N2 (Relatedness)", "S2-N3 (Mastery)"]
idxA_sorted = sorted(idxA)
idxB_sorted = sorted(idxB)
label_by_index = {vi: labels_S1[k] for k, vi in enumerate(idxA_sorted)}
label_by_index.update({vi: labels_S2[k] for k, vi in enumerate(idxB_sorted)})

# =========================
# Define a continuous packet route
# =========================
# A Hamiltonian cycle on tetra A (visit 4 vertices, close the loop)
cycle_A = [idxA_sorted[0], idxA_sorted[1], idxA_sorted[2], idxA_sorted[3], idxA_sorted[0]]

# Bridge to tetra B via IC: from last A vertex -> IC -> nearest B vertex
last_A = cycle_A[-1]
# Choose the B vertex closest to last_A (Euclidean)
dists = np.linalg.norm(V[idxB_sorted] - V[last_A], axis=1)
start_B = idxB_sorted[int(np.argmin(dists))]

# A Hamiltonian cycle on tetra B, starting at start_B
others_B = [v for v in idxB_sorted if v != start_B]
cycle_B = [start_B] + others_B + [start_B]

# Return via IC back to A start for looping
route_vertices = cycle_A + ["IC"] + [start_B] + cycle_B[1:] + ["IC", cycle_A[0]]
# Convert to 3D point list with IC inserted where needed
def vcoord(tag):
    return IC if tag == "IC" else V[tag]

route_points = [vcoord(v) for v in route_vertices]

# =========================
# Build animated path samples
# =========================
# Interpolate along each consecutive pair to build a continuous polyline
samples = []
per_seg = 30  # resolution per segment (increase for smoother motion)
for p0, p1 in zip(route_points[:-1], route_points[1:]):
    if np.allclose(p0, p1):
        continue
    seg = interp_segment(np.array(p0), np.array(p1), n=per_seg)
    samples.append(seg)
# Add the very final endpoint
samples.append(np.array([route_points[-1]]))
samples = np.vstack(samples)  # [N, 3]

# Optional tail length (in samples)
tail = 90  # set 0 for no trail

# =========================
# Static shape traces (cube + star-tetra edges + labels + EC ring)
# =========================
# Cube edges
cx, cy, cz = segments_to_lines(V, E_cube)
tr_cube = go.Scatter3d(x=cx, y=cy, z=cz, mode="lines",
                       line=dict(width=2), opacity=0.25,
                       name="Cube edges", hoverinfo="skip")

# Tetra A edges
ax_, ay_, az_ = segments_to_lines(V, E_A)
tr_tetA = go.Scatter3d(x=ax_, y=ay_, z=az_, mode="lines",
                       line=dict(width=6),
                       name="Tetrahedron A (S1)")

# Tetra B edges
bx_, by_, bz_ = segments_to_lines(V, E_B)
tr_tetB = go.Scatter3d(x=bx_, y=bz_, z=bz_, mode="lines")  # placeholder, fix below
tr_tetB = go.Scatter3d(x=bx_, y=by_, z=bz_, mode="lines",
                       line=dict(width=6),
                       name="Tetrahedron B (S2)")

# Vertex markers with labels
tr_vertices = go.Scatter3d(
    x=V[:,0], y=V[:,1], z=V[:,2],
    mode="markers+text",
    text=[label_by_index.get(i, f"v{i}") for i in range(len(V))],
    textposition="top center",
    marker=dict(size=4),
    name="S1/S2 vertices",
    hovertemplate="%{text}<br>(%{x}, %{y}, %{z})<extra></extra>"
)

# IC at centroid
tr_ic = go.Scatter3d(
    x=[IC[0]], y=[IC[1]], z=[IC[2]],
    mode="markers+text",
    text=["IC"],
    textposition="top center",
    marker=dict(size=7),
    name="IC (Internal Coherence)",
    hovertemplate="IC (Internal Coherence)<extra></extra>"
)

# EC: static blue circle in z=0 plane
r = np.sqrt(2) * a * 1.12
theta = np.linspace(0, 2*np.pi, 361)
EC_circle = np.column_stack([r*np.cos(theta), r*np.sin(theta), np.zeros_like(theta)])
tr_ec = go.Scatter3d(x=EC_circle[:,0], y=EC_circle[:,1], z=EC_circle[:,2],
                     mode="lines",
                     line=dict(width=4),
                     name="EC (External Coherence)",
                     hoverinfo="skip")

# EC label on ring
ec_label_idx = 40
tr_ec_label = go.Scatter3d(
    x=[EC_circle[ec_label_idx,0]],
    y=[EC_circle[ec_label_idx,1]],
    z=[EC_circle[ec_label_idx,2]],
    mode="text",
    text=["EC"],
    showlegend=False,
    hoverinfo="skip"
)

# =========================
# Animated packet traces
# =========================
# Initial state
p0 = samples[0]
tr_packet = go.Scatter3d(
    x=[p0[0]], y=[p0[1]], z=[p0[2]],
    mode="markers",
    marker=dict(size=6),
    name="Packet"
)
# Trail (path travelled)
tr_trail = go.Scatter3d(
    x=[p0[0]], y=[p0[1]], z=[p0[2]],
    mode="lines",
    line=dict(width=5),
    name="Trail",
    hoverinfo="skip"
)

# Frames
frames = []
for i in range(len(samples)):
    # trail window
    j0 = max(0, i - tail)
    xs = samples[j0:i+1, 0]
    ys = samples[j0:i+1, 1]
    zs = samples[j0:i+1, 2]
    frames.append(go.Frame(
        name=str(i),
        data=[
            go.Scatter3d(x=[samples[i,0]], y=[samples[i,1]], z=[samples[i,2]]),  # packet
            go.Scatter3d(x=xs, y=ys, z=zs)                                        # trail
        ],
        traces=[6, 7]  # indices (packet, trail) in the final fig.data list
    ))

# =========================
# Controls
# =========================
play_btn = dict(
    label="▶ Play",
    method="animate",
    args=[None, {"fromcurrent": True,
                 "frame": {"duration": 25, "redraw": True},
                 "transition": {"duration": 0}}]
)
pause_btn = dict(
    label="⏸ Pause",
    method="animate",
    args=[[None], {"mode": "immediate",
                   "frame": {"duration": 0, "redraw": False},
                   "transition": {"duration": 0}}]
)

updatemenus = [dict(
    type="buttons",
    showactive=False,
    x=0.5, y=1.08, xanchor="center",
    buttons=[play_btn, pause_btn]
)]

layout = go.Layout(
    title="Packet Flow over the Star Tetrahedron • IC bridge • Continuous Cycle",
    scene=dict(
        xaxis=dict(visible=False),
        yaxis=dict(visible=False),
        zaxis=dict(visible=False),
        aspectmode="data",
    ),
    legend=dict(x=0.01, y=0.99),
    margin=dict(l=0, r=0, t=60, b=0),
    updatemenus=updatemenus
)

# =========================
# Figure
# =========================
fig = go.Figure(
    data=[tr_ec, tr_ec_label, tr_cube, tr_tetA, tr_tetB, tr_vertices, tr_packet, tr_trail, tr_ic],
    layout=layout,
    frames=frames
)

# A touch of colour to distinguish structures
fig.update_traces(selector=dict(name="Tetrahedron A (S1)"), line=dict(color="#1f77b4"))
fig.update_traces(selector=dict(name="Tetrahedron B (S2)"), line=dict(color="#ff7f0e"))
fig.update_traces(selector=dict(name="Cube edges"), line=dict(color="#888"))
fig.update_traces(selector=dict(name="EC (External Coherence)"), line=dict(color="#1976d2"))
fig.update_traces(selector=dict(name="Packet"), marker=dict(color="#111"))
fig.update_traces(selector=dict(name="Trail"), line=dict(color="#111"))

fig.show()

# Mapping log
print("Vertex → Label mapping:")
for i in range(len(V)):
    lab = label_by_index.get(i, f"v{i}")
    print(f"{i}: {lab}  coords={tuple(V[i])}")
print("IC:", tuple(IC))
print("Route length (samples):", len(samples))
```
